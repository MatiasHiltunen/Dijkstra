```python
import gymnasium as gym
from ShortestPathEnvDijkstra import ShortestPathEnv
import numpy as np
from IPython.display import clear_output
import time
```

```python
# Create an instance of the environment
env = ShortestPathEnv(random_start=False, random_end=False)

# Reset the environment
observation = env.reset()
```

```python
env._add_travel_times()
```

```python
env.render()
```

```python
env.reset()
total_steps = 100
shortest_path = env.get_shortest_path(mode='travel_time')
for step in shortest_path[1:]:
    action = step
    next_observation, reward, done, truncated, info = env.step(action)
    env.render()
    if done:
        print("Episode finished after {} timesteps".format(step+1))
        time_taken = env.time_travelled * 3600
        print(f"Time taken: {int(time_taken // 60)}m{int(time_taken% 60)}s, Distance travelled: {int(env.distance_travelled)}")
        break
    print("Step: ", step+1, "Observation: ", observation, "Action: ", action, "Reward: ", reward)
    clear_output(wait=True)
env.close()
```
